```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
train = pd.read_csv('train.csv',index_col=0)
test = pd.read_csv('test.csv',index_col=0)
```

```python
train['Cabin_new'] = train['Cabin'].apply(lambda x: x[0] if type(x) == str else x)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train['Age_new'] = train['Age'].fillna(train['Age'].mean)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train.groupby('Embarked').count()
```

```python
train['Embarked_new'] = train['Embarked'].fillna('S')
```

```python
train.isnull().sum()
```

## generate dummies

```python
sex_dummy = pd.get_dummies(train['Sex'])
```

```python
train = pd.concat([train, sex_dummy],axis=1)
train.drop('Sex',axis=1,inplace = True)
train.head()
```

```python
embarked_dummy = pd.get_dummies(train['Embarked_new'])
```

```python
train = pd.concat([train, embarked_dummy],axis=1)
train.drop('Embarked_new',axis=1,inplace = True)
train.head()
```

```python
cabin_dummy = pd.get_dummies(train['Cabin_new'])
```

```python
train = pd.concat([train, cabin_dummy],axis=1)
train.drop('cabin_new',axis=1,inplace = True)
train.head()
```

```python
train.drop(['Cabin','Embarked'],axis=1,inplace = True)
```

```python
train.drop(['Cabin_new'],axis=1,inplace = True)
```

```python
train.drop(['Age'],axis=1,inplace = True)
```

```python
train.drop(['Age_new'],axis=1,inplace = True)
```

```python
train.info()
```

```python
train.drop(['Name','Ticket'],axis=1,inplace = True)
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X= train.drop(['PassangerId','Survived'],axis=1),y=train['Survived'])
```

```python
prediction = rf.predict(train.drop(['PassangerId','Survived'],axis=1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(train['Survived'],prediction)
```

```python
print(classification_report(train['Survived'],prediction))
```
